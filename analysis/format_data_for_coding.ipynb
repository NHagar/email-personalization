```python
import ast

import pandas as pd
```

```python
df = pd.read_csv("../data/responses_full.csv")
```

```python
df['history'] = df['history'].apply(ast.literal_eval)
df['choices'] = df['choices'].apply(ast.literal_eval)
```

```python
df = df.drop_duplicates(subset=["user_id"])
```

```python
items = {
    "radio_0": [],
    "radio_1": [],
    "radio_2": [],
    "radio_3": [],
    "radio_4": [],
}
```

```python
def process_choices(choices):
    for radio, choice in choices.items():
        for headline, _ in choice["options"].items():
            items[radio].append(headline)
```

```python
df.choices.apply(process_choices)
```

```python
def cleanup_items(item_list):
    return [i.replace("#### ", "").replace("**", "").replace("\n", ". ") for i in item_list]
```

```python
items = {k: list(set(v)) for k, v in items.items()}
```

```python
items = {k: cleanup_items(v) for k, v in items.items()}
```
